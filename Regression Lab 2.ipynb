```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Attendance.csv")
```

```python
data.head()
```

```python
x=data['MSE']
y=data['ESE']
sns.scatterplot(x,y)
```

```python
beta0=0
beta1=0
alpha=0.01
count=10000
n=float(len(x))
```

```python
for i in range(count):
    Ybar=beta1*x+beta0
    beta1=beta1-(alpha/n)*sum(x*(Ybar-y))
    beta0=beta0-(alpha/n)*sum(Ybar-y)
    
    print(beta0,beta1)
    
```

```python
Ybar=beta1*x+beta0
plt.scatter(x,y)
plt.plot([min(x), max(x)],[min(Ybar),max(Ybar)],color='r') #regression line
plt.show()
```

```python
import math
def RES(y_true,y_predicted):
    y_true=np.array(y_true)
    y_predicted=np.array(y_predicted)
    RSS=np.sum(np.square(y_true-y_predicted))
    
    rse=math.sqrt(RSS / (len(y_true)-2))
    return rse

res=RES(data['ESE'],Ybar)
print(res)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x=np.array(data['MSE']).reshape(-1,1)
y=np.array(data['ESE']).reshape(-1,1)
lr=LinearRegression()
lr.fit(x,y)
print(lr.coef_)
print(lr.intercept_)

yp=lr.predict(x)
rse=RES(y,yp)

print(rse)

```
